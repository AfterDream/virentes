### Convert SNP data to FS   
For large data sets fs will be counted from many SNPs and so is easiest to read in from a file. In this case I convert a `.loci` output file with thousands of RADseq loci from an anlysis in _pyRAD_ into the SNP format used by _dadi_, and then extract the site frequency spectrum from that file. 

```python
## define focal populations and the outgroups
cuba        = ['CUCA4','CUVN10','CUSV6']
centralA    = ['CRL0030','CRL0001','HNDA09','BZBB1','MXSA3017']
virginiana  = ['FLSF33','FLBA140','SCCU3']   ## exclude western virg sample
geminata    = ["FLCK18","FLSF54","FLWO6","FLAB109"]
minima      = ["FLSF47","FLMO62","FLSA185","FLCK216"]
outgroup    = ["DO","DU","EN","AR","CH","NI","HE"]

florida = virginiana+geminata+minima
```

```python
## a function to select most common element in a list
def most_common(L):
    return max(itt.groupby(sorted(L)), key=lambda(x, v):(len(list(v)),
                                                     -L.index(x)))[0]

## a function to get alleles from ambiguous bases
def unstruct(amb):
    " returns bases from ambiguity code"
    D = {"R":["G","A"],
         "K":["G","T"],
         "S":["G","C"],
         "Y":["T","C"],
         "W":["T","A"],
         "M":["C","A"],
         "A":["A","A"],
         "T":["T","T"],
         "G":["G","G"],
         "C":["C","C"],
         "N":["N","N"],
         "-":["-","-"]}
    return D.get(amb)
```

```python
## parse the loci
locifile = open("/home/deren/Dropbox/oaks.testdadi.loci")
loci = locifile.read().strip().split("|")
```

```python
## Finds the trinucleotide reference outgroup allele and ingroup reference allele
## from the most common alleles in each group at each SNP. Writes these to a file. 

## outfile name
outfile = open("oaks.dadisnps", "w")

## print header
print >>outfile, "\t".join(["Ingroup","Outgroup",
                         "Allele1","florida","centralA","cuba",
                         "Allele2","florida","centralA","cuba",
                         "locus","site"])

## only examine loci w/ at least 1 outgroup sample
## and at least two inds from each focal populations
for locN, loc in enumerate(loci):
    names = [i.strip().split(" ")[0] for i in loc.strip().split("\n")[:-1]]  
    if len(set([">"+i for i in outgroup]).intersection(names)) > 1:
        if len(set([">"+i for i in florida]).intersection(names)) >= 2:
            if len(set([">"+i for i in centralA]).intersection(names)) >= 2:
                if len(set([">"+i for i in cuba]).intersection(names)) >= 2:
                    ## only examine loci that are variable in the ingroup samples
                    dat = [i.strip().split(" ")[-1] for i in loc.strip().split("\n")[:-1]]
                    snps = loc.strip().split("\n")[-1][14:]
                    ihits = [names.index(">"+i) for i in cuba+centralA+florida \
                              if ">"+i in names]
                    ohits = [names.index(">"+i) for i in outgroup \
                              if ">"+i in names]
                    for site,char in enumerate(snps):
                        i1 = i2 = i3 = o1 = o2 = o3 = ""
                        if char in["-","*"]:
                            ## get ingroup common (reference) alleles
                            i1 = [dat[k][site-1] if dat[k][site-1] not in ["N"] else "-" for k in ihits]
                            i2 = [dat[k][site] for k in ihits if dat[k][site] not in ["N",'-']]
                            i3 = [dat[k][site+1] if dat[k][site+1] not in ["N"] else "-" for k in ihits]
                            ## get outgroup common (reference) alleles
                            o1 = [dat[k][site-1] if dat[k][site-1] not in ["N"] else "-" for k in ohits]
                            o2 = [dat[k][site] for k in ohits if dat[k][site] not in ["N",'-']]
                            o3 = [dat[k][site+1] if dat[k][site+1] not in ["N"] else "-" for k in ohits]
                        ## if all sites have data
                        if all([i for i in [i1,i2,i3,o1,o2,o3]]):
                            ## if site is variable in the ingroups
                            if len(set(i2)) > 1:
                                allele1 = most_common(list(itt.chain(*[unstruct(i) for i in i2])))
                                temp = [i for i in list(itt.chain(*[unstruct(i) for i in i2])) if i!=allele1]
                                allele2 = most_common(temp)                            
                                a= most_common(list(itt.chain(*[unstruct(i) for i in i1])))
                                b= most_common(list(itt.chain(*[unstruct(i) for i in i2])))
                                c= most_common(list(itt.chain(*[unstruct(i) for i in i3])))
                                d= most_common(list(itt.chain(*[unstruct(i) for i in o1])))
                                e= most_common(list(itt.chain(*[unstruct(i) for i in o2])))
                                f= most_common(list(itt.chain(*[unstruct(i) for i in o3])))
                                ## get florida alleles
                                fl = [names.index(">"+z) if ">"+z in names else '-' for z in florida]
                                fldat = list(itt.chain(*[unstruct(dat[z][site]) for z in fl if z not in ["N","-"]]))
                                fl1 = fldat.count(allele1)
                                fl2 = fldat.count(allele2)
                                ## get centraA alleles
                                ca = [names.index(">"+z) if ">"+z in names else '-' for z in centralA]
                                cadat = list(itt.chain(*[unstruct(dat[z][site]) for z in ca if z not in ["N","-"]]))
                                ca1 = cadat.count(allele1)
                                ca2 = cadat.count(allele2)
                                ## get cuba alleles 
                                cu = [names.index(">"+z) if ">"+z in names else '-' for z in cuba]
                                cudat = list(itt.chain(*[unstruct(dat[z][site]) for z in cu if z not in ["N","-"]]))
                                cu1 = cudat.count(allele1)
                                cu2 = cudat.count(allele2)
                                ## print 
                                print >>outfile, "\t".join(map(str,[a+b+c, d+e+f,
                                                                 allele1, fl1, ca1, cu1,
                                                                 allele2, fl2, ca2, cu2,
                                                                 str(locN), str(site)]))
                                
outfile.close()
```

# Define four different _IM_-type models  
Models `IM1` and `IM2` are isolation-migration models for a 3 population tree in which Cuba is derived from on lineage or the other. The `IMadmix` and `admix` models have the Cuban population form as an instantaneous fusion of the other two populations, after which it either exchanges some gene flow with the parent populations, or remains isolated. 

![](https://dl.dropboxusercontent.com/u/2538935/PyRAD_TUTORIALS/tutorial_dadi/modelsFig.png)

### Model 1: Island conversion model

```python
## scenario ((florida,cuba),centralA);

def IM_split1(params, ns, pts):
    ## parse params
    N,T2,T1,m12,m13,m21,m23,m31,m32 = params
    
    ## create a search grid
    xx = dadi.Numerics.default_grid(pts)
    
    ## create ancestral pop
    phi = dadi.PhiManip.phi_1D(xx)
    
    ## split ancestral pop into two species
    phi = dadi.PhiManip.phi_1D_to_2D(xx,phi)
    
    ## allow drift to occur along each of these branches
    phi = dadi.Integration.two_pops(phi, xx, T2, nu1=N, nu2=N, m12=0., m21=0.)
    
    ## split pop1 into pops 1 and 3
    phi = dadi.PhiManip.phi_2D_to_3D_split_1(xx, phi)

    ## allow drift and migration to occur along these branches
    phi = dadi.Integration.three_pops(phi, xx, T1,
                                       nu1=N, nu2=N, nu3=N,  
                                       m12=m12, m13=m13, 
                                       m21=m21, m23=m23,
                                       m31=m31, m32=m32)
    ## simulate the fs
    fs = dadi.Spectrum.from_phi(phi,ns,(xx,xx,xx), 
                                pop_ids=['florida', 'centralA', 'cuba'])
    return fs
```

### Model 2: Island conversion model

```python
## scenario ((centralA,cuba),florida);

def IM_split2(params, ns, pts):
    ## parse params
    N,T2,T1,m12,m13,m21,m23,m31,m32 = params
    
    ## create a search grid
    xx = dadi.Numerics.default_grid(pts)
    
    ## create ancestral pop
    phi = dadi.PhiManip.phi_1D(xx)
    
    ## split ancestral pop into two species
    phi = dadi.PhiManip.phi_1D_to_2D(xx,phi)
    
    ## allow drift to occur along each of these branches
    phi = dadi.Integration.two_pops(phi, xx, T2, nu1=N, nu2=N, m12=0., m21=0.)
    
    ## split pop2 into pops 2 and 3
    phi = dadi.PhiManip.phi_2D_to_3D_split_2(xx, phi)

    ## allow drift and migration to occur along these branches
    phi = dadi.Integration.three_pops(phi, xx, T1,
                                      nu1=N, nu2=N, nu3=N,  
                                      m12=m12, m13=m13, 
                                      m21=m21, m23=m23,
                                      m31=m31, m32=m32)
    ## simulate the fs
    fs = dadi.Spectrum.from_phi(phi,ns,(xx,xx,xx), 
                                pop_ids=['florida', 'centralA', 'cuba'])
    return fs
```

### Model 3: admixture and gene flow model

```python
## scenario hybrid speciation forms cuba, and subsequent gene flow

def IM_admix(params, ns, pts):
    ## parse params
    N,T2,T1,m12,m13,m21,m23,m31,m32,f = params
    
    ## create a search grid
    xx = dadi.Numerics.default_grid(pts)
    
    ## make ancestral pop that splits into two
    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.PhiManip.phi_1D_to_2D(xx,phi)
    
    ## allow drift to occur along each of these branches
    phi = dadi.Integration.two_pops(phi, xx, T2, nu1=N, nu2=N, m12=0., m21=0.)
    
    ## create pop 3 from a mixture of 1 and 2
    phi = dadi.PhiManip.phi_2D_to_3D_admix(phi, f, xx, xx, xx)

    ## allow drift and migration to occur along these branches
    ## cuba population shrinks in size after divergence
    phi = dadi.Integration.three_pops(phi, xx, T1,
                                      nu1=N, nu2=N, nu3=N,  
                                      m12=m12, m13=m13, 
                                      m21=m21, m23=m23,
                                      m31=m31, m32=m32)
    ## simulate the fs
    fs = dadi.Spectrum.from_phi(phi,ns,(xx,xx,xx),
                                pop_ids=['florida', 'centralA', 'cuba'])
    return fs
```

### Model 4: Hybrid speciation -- admixture and isolation model  
Currently exploring with a parameter to shrink the cuba population size

```python
## scenario hybrid speciation forms cuba

def admix(params, ns, pts):
    ## parse params
    N,N_Cuba,T2,T1,f = params
    
    ## create a search grid
    xx = dadi.Numerics.default_grid(pts)
    
    ## make ancestral pop that splits into two
    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.PhiManip.phi_1D_to_2D(xx,phi)
    
    ## allow drift to occur along each of these branches
    phi = dadi.Integration.two_pops(phi, xx, T2, nu1=N, nu2=N, m12=0., m21=0.)
    
    ## create pop 3 from a mixture of 1 and 2
    phi = dadi.PhiManip.phi_2D_to_3D_admix(phi, f, xx, xx, xx)

    ## allow drift to occur along these branches
    N_Cuba_func = lambda t : N * (N_Cuba/N) ** (t/T1)
    phi = dadi.Integration.three_pops(phi, xx, T1,
                                      nu1=N, nu2=N, nu3=N_Cuba_func, 
                                      m12=0.0, m13=0.0, 
                                      m21=0.0, m23=0.0,
                                      m31=0.0, m32=0.0)
    ## simulate the fs
    fs = dadi.Spectrum.from_phi(phi,ns,(xx,xx,xx),
                                pop_ids=['florida','centralA','cuba'])
    return fs
```

### Function for model fitting

```python
## Function to repeat optimization N times from different starting params
def Fit(reps, maxiters, model, params, ns, pts_l, uppers, lowers, log=0):
    results = []
    ## create extrapolating func
    if log:
        Func = dadi.Numerics.make_extrap_log_func(model)
    else:
        Func = dadi.Numerics.make_extrap_func(model)
    ## get model AFS
    model_AFS = Func(params,ns,pts_l)
    for rep in range(reps):
        ## perturb start values
        p0 = dadi.Misc.perturb_params(params, fold=1, 
                                      upper_bound=uppers,
                                      lower_bound=lowers)
        ## optimize params
        args = [p0, fs, Func, pts_l, lowers, uppers, 0, maxiters]
        if log:
            popt = dadi.Inference.optimize_log(*args)
        else:
            popt = dadi.Inference.optimize(*args)
        results.append(popt)
    return results
```
